```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
from torchvision.datasets import ImageFolder
from tqdm.autonotebook import tqdm as tqdm_jupyter
from tqdm import tqdm as tqdm_bash
from PIL import Image
import dask
```

```python
train_dir = "../data/train"
test_dir = "../data/test"
train_dset = ImageFolder(root=train_dir)

pbar_type = "jupyter"
```

```python
images = []
labels = []

if pbar_type == "bash":
    progress = tqdm_bash(train_dset, total=len(train_dset), unit=" image(s)", position=0, leave=True)
if pbar_type == "jupyter":
    progress = tqdm_jupyter(train_dset, total=len(train_dset), unit=" image(s)")


for pair in progress:
    im = pair[0]
    im.thumbnail((224,224), Image.ANTIALIAS)
    images.append(np.array(im))
    labels.append(pair[1])
```
